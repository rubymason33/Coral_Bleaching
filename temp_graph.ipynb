```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
```

```python
conn = conn  = sqlite3.connect('temp.db')
if conn:
    print('success')
else:
    print('failure')
```

```python
query = pd.read_sql_query('''SELECT year, sum(temp) AS carbon_ppm FROM globaltemp
	                            GROUP BY year;''', conn)
carbon_year = pd.DataFrame(query)
carbon_year = carbon_year.drop([66])
```

```python
plt.figure()
plt.plot(carbon_year['year'], carbon_year['carbon_ppm'])
plt.show()
```
